## Plotting positions

```python
from scipy.io import loadmat
matlab_file = 'labsea_66.44W_45.68N_5min.mat'
mat_data = loadmat(os.path.join('..', data_folder, matlab_file))
```

```python
bathy_data = mat_data['bathy']
```

```python
lat = bathy_data['lat'][0][0].flatten()
lon = bathy_data['lon'][0][0].flatten()
dep = bathy_data['depth'][0][0]
```

```python
[len(lat), len(lon), dep.shape]
```

```python
fig, ax = plt.subplots()
ax.contour(lon, lat, dep)
glider_path.plot(x='longitude',y='latitude', color='r', ax=ax)
plt.show()

# Fix aspect ratio
```
